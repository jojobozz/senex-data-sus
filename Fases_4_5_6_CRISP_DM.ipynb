```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url = 'https://github.com/ITA-LOW/senex-data-sus/blob/main/data_frame_p%C3%B3s_ETL.csv?raw=true'
df = pd.read_csv(url)
df.drop(columns=['Unnamed: 0','ESTADO','ANO'], inplace=True)
```

```python
X = df.iloc[:, :-1].values
y = pd.DataFrame(df.iloc[:, 4].values,columns=['CAUSABAS'])

y
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y = y.apply(le.fit_transform)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state=99)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier

mlp = MLPClassifier(hidden_layer_sizes=(10,10,10),max_iter=3000,random_state=171)
mlp.fit(X_train, y_train.values.ravel())
```

```python
#essa celula eh pra testar se outras arquiteturas performam melhor
from sklearn.model_selection import cross_val_score

test_arch = [(1,3,2),(1,5,7),(3,3),(9,5,1),(15,11,5)]

for arch in test_arch:
    print(f"Testando arquitetura {arch}")
    model = MLPClassifier(hidden_layer_sizes=arch, random_state=1, max_iter=2000)
    accuracy = cross_val_score(model, X_train, y_train.values.ravel(), cv = 10)
    print(accuracy)
    print("Acuracia media: ",accuracy.mean())
```
